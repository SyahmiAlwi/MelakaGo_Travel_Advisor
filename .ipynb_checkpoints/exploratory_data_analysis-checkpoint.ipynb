```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("traffic_with_weather_cleaned.csv", parse_dates=["datetime"])
holidays_df = pd.read_csv("malaysia_public_holiday_2024.csv", parse_dates=["date"])
```

```python
# Daily trend
daily_traffic = df.resample("D", on="datetime")["total_traffic"].sum()

plt.figure()
daily_traffic.plot(title="Daily Total Traffic (2024)")
plt.xlabel("Date")
plt.ylabel("Total Traffic")
plt.show()
```

```python
# Diurnal pattern
df["hour"] = df["datetime"].dt.hour

plt.figure()
plt.boxplot([df[df.hour==h]["total_traffic"] for h in range(24)], showfliers=False)
plt.title("Diurnal Traffic Pattern")
plt.xlabel("Hour of Day")
plt.ylabel("Total Traffic")
plt.show()
```

```python
# 5. Weekend vs weekday
df["is_weekend"] = df["is_weekend"].astype(bool)
weekday_avg = df[~df["is_weekend"]].groupby("hour")["total_traffic"].mean()
weekend_avg = df[df["is_weekend"]].groupby("hour")["total_traffic"].mean()

plt.figure()
plt.plot(weekday_avg.index, weekday_avg.values, label="Weekday")
plt.plot(weekend_avg.index, weekend_avg.values, label="Weekend")
plt.title("Weekday vs Weekend Hourly Traffic")
plt.xlabel("Hour")
plt.ylabel("Avg Traffic")
plt.legend()
plt.show()
```

```python
vehicle_types = ["motorcycle", "car", "truck", "bus"]

for vt in vehicle_types:
    bend_col = f'bendahara_{vt}'
    tem_col  = f'temenggong_{vt}'
    tot_col  = f'total_{vt}'
    df[tot_col] = df[bend_col] + df[tem_col]
    
df.head()
```

```python
# 6. Correlation matrix
corr_matrix = df[["total_traffic", "total_motorcycle", "total_car", "total_truck", "total_bus", "temperature_2m","relative_humidity_2m","windspeed_10m"]].corr()

plt.figure()
plt.matshow(corr_matrix, fignum=1)
plt.colorbar()
plt.title("Correlation Matrix", pad=20)
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix)), corr_matrix.index)
plt.show()
```

```python
# 7. Traffic vs weather scatter
scatter_data = df.sample(1000)  # sample for clarity

plt.figure()
plt.scatter(scatter_data["temperature_2m"], scatter_data["total_traffic"], alpha=0.5)
plt.title("Traffic vs Temperature")
plt.xlabel("Temperature (°C)")
plt.ylabel("Total Traffic")
plt.show()
```

```python
plt.figure()
plt.scatter(scatter_data["windspeed_10m"], scatter_data["total_traffic"], alpha=0.5)
plt.title("Traffic vs Windspeed")
plt.xlabel("Wind Speed (m/s)")
plt.ylabel("Total Traffic")
plt.show()
```

```python
# Build the mask for “applies in Malacca”
mask_apply = (
    # (a) “All” without excluding Malacca
    (holidays_df.states_applicable.str.contains(r"\bAll\b", regex=True)) &
    (~holidays_df.states_applicable.str.contains(r"except.*\bMalacca\b", regex=True))
) | (
    # (b) Explicitly lists “Melaka”
    holidays_df.states_applicable.str.contains(r"\bMalacca\b", regex=True)
)

# Filter
malacca_holidays = holidays_df[mask_apply].copy()

# Create holiday indicator
df["date_only"] = df["datetime"].dt.date
malacca_holiday_dates = set(malacca_holidays["date"].dt.date)
holiday_dates = set(holidays_df["date"].dt.date)
df["is_holiday_mlk"] = df["date_only"].isin(malacca_holiday_dates)
df["is_holiday"] = df["date_only"].isin(holiday_dates)
```

```python
# Traffic on holidays vs. non-holidays
avg_traffic_by_holiday = df.groupby("is_holiday")["total_traffic"].mean().rename({False: "Non-Hol", True: "Holiday"})
print("Average Total Traffic:\n", avg_traffic_by_holiday)

# Average traffic
avg_by_hol = df.groupby("is_holiday_mlk")["total_traffic"].mean()
print("Average traffic (Non‐holiday vs Holiday in Malacca):")
print(avg_by_hol.rename({False: "Non-Holiday", True: "Holiday"}))
```

```python
# Boxplot of traffic distribution
plt.figure()
df.boxplot(column="total_traffic", by="is_holiday", showfliers=False)
plt.title("Traffic on Holidays vs. Non-Holidays")
plt.suptitle("")
plt.xlabel("Is Holiday")
plt.ylabel("Total Traffic")
plt.show()
```

```python
# Boxplot
plt.figure()
df.boxplot(column="total_traffic", by="is_holiday_mlk", showfliers=False)
plt.suptitle("")  # remove default
plt.xlabel("Holiday in Malacca")
plt.ylabel("Total Traffic")
plt.title("Traffic Distribution: Malacca Holidays vs Non-Holidays")
plt.show()
```

```python
# Hourly pattern: holidays vs non-holidays
df["hour"] = df["datetime"].dt.hour
hourly_hol = df[df.is_holiday].groupby("hour")["total_traffic"].mean()
hourly_non = df[~df.is_holiday].groupby("hour")["total_traffic"].mean()

plt.figure()
plt.plot(hourly_non.index, hourly_non.values, label="Non-Holiday")
plt.plot(hourly_hol.index, hourly_hol.values, label="Holiday")
plt.title("Hour-of-Day Traffic: Holidays vs. Non-Holidays")
plt.xlabel("Hour")
plt.ylabel("Avg Total Traffic")
plt.legend()
plt.show()
```

```python
df["hour"] = df["datetime"].dt.hour

hourly_hol = df[df.is_holiday_mlk].groupby("hour")["total_traffic"].mean()
hourly_non = df[~df.is_holiday_mlk].groupby("hour")["total_traffic"].mean()

plt.figure()
plt.plot(hourly_non.index, hourly_non.values, label="Non-Holiday")
plt.plot(hourly_hol.index, hourly_hol.values, label="Holiday")
plt.xlabel("Hour of Day")
plt.ylabel("Avg Traffic")
plt.title("Malacca: Hourly Traffic on Holidays vs Non-Holidays")
plt.legend()
plt.show()
```

```python
df["weekday"] = df["datetime"].dt.day_name().str[:3]  # Mon, Tue, ...
pivot = df.pivot_table(
    index="weekday", columns="hour", values="total_traffic", aggfunc="mean"
).reindex(["Mon","Tue","Wed","Thu","Fri","Sat","Sun"])

plt.figure(figsize=(12,5))
sns.heatmap(pivot, cmap="YlOrRd", linewidths=.5)
plt.title("Average Traffic by Weekday and Hour (Malacca 2024)")
plt.xlabel("Hour of Day")
plt.ylabel("Weekday")
plt.show()
```

```python
# Use total_traffic vs. a threshold (e.g. 90th percentile) to flag “traffic jams”
jam_threshold = df["total_traffic"].quantile(0.80)
df["is_jam"] = df["total_traffic"] >= jam_threshold
```

```python
# Group by temperature bins or wind‐speed bins and compute jam frequency

# Temperature bins
df["temp_bin"] = pd.cut(df["temperature_2m"], bins=[20,25,30,35,40], labels=["20-25","25-30","30-35","35+"])
jam_by_temp = df.groupby("temp_bin", observed=True)["is_jam"].mean()

# Plot
jam_by_temp.plot(kind="bar", title="Jam Rate by Temperature Bin")
```

### Interpretation: A rising jam rate in “30–35°C” suggests heat-related slowdowns.

```python
mapping = pd.read_csv("weather_code_mapping.csv")

# Compute jam count by weathercode
jam_stats = (
    df
    .groupby("weathercode")["is_jam"]
    .sum()
    .reset_index(name="jam_count")
)

# Merge in the readable descriptions
jam_table = jam_stats.merge(
    mapping,
    left_on="weathercode",
    right_on="code",
    how="left"
).loc[:, ["weathercode", "description", "jam_count"]]

jam_rate = df.groupby("weathercode")["is_jam"].mean().reset_index(name="jam_rate")
jam_table = jam_table.merge(jam_rate, on="weathercode")

print(jam_table)
```

```python
# Plot histogram of jam_count
plt.figure()
plt.bar(jam_table['weathercode'].astype(str), jam_table['jam_count'])
plt.title("Jam Count by Weather Code")
plt.xlabel("Weather Code")
plt.ylabel("Jam Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Vehicle‐type risk at certain hours

# Calculate vehicle type shares
df["car_share"] = df["total_car"] / df["total_traffic"]
df["motorcycle_share"] = df["total_motorcycle"] / df["total_traffic"]
df["truck_share"] = df["total_truck"] / df["total_traffic"]
df["bus_share"] = df["total_bus"] / df["total_traffic"]

# Calculate the average share of each mode by hour
hourly_shares = df.groupby('hour')[
    ['car_share', 'motorcycle_share', 'truck_share', 'bus_share']
].mean()

# Plot all vehicle-type shares
plt.figure()
for col in hourly_shares.columns:
    plt.plot(hourly_shares.index, hourly_shares[col], label=col)
plt.title("Hourly Vehicle-Type Mode Shares")
plt.xlabel("Hour of Day")
plt.ylabel("Share of Total Traffic")
plt.xticks(range(0, 24))
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Identify peak vs. non-peak hours

# Cluster hours by average traffic

# Use k-means on the 24 hourly means to divide into three groups

from sklearn.cluster import KMeans
import numpy as np

hourly_mean = df.groupby("hour")["total_traffic"].mean().to_frame()
kmeans = KMeans(n_clusters=3, random_state=0).fit(hourly_mean)
hourly_mean["cluster"] = kmeans.labels_

# Map clusters to labels by their centroid order
centroids = kmeans.cluster_centers_.flatten()
order = np.argsort(centroids)  # [off-peak, shoulder, peak]
label_map = {order[0]:"Off-Peak", order[1]:"Shoulder", order[2]:"Peak"}
hourly_mean["label"] = hourly_mean["cluster"].map(label_map)

print(hourly_mean[["total_traffic","label"]])
```

```python
# Apply labels back to dataset
df = df.merge(hourly_mean[["label"]], on="hour", how="left")
df.rename(columns={"label":"peak_category"}, inplace=True)
df
```

```python
# Extract time-based features
df["hour"] = df["datetime"].dt.hour
df["day_of_week"] = df["datetime"].dt.day_name()
df["month"] = df["datetime"].dt.month_name()

# Display head to verify
df[["datetime", "hour", "day_of_week", "month"]].head()
```

```python
# Drop a column in place
df.drop(labels=['temp_bin', 'weekday'], axis=1, inplace=True)
print(df.columns)
print(df.info())
df.describe().T
```
